```python
import discretize
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mesh = discretize.TensorMesh([100, 100])
ux = np.sin(mesh.gridEx[:, 0]) * np.cos(mesh.gridEx[:, 1])
uy = np.cos(10*mesh.gridEy[:, 0]) * np.cos(10*mesh.gridEy[:, 1])
u = np.hstack([ux, uy])
```

```python
xlim = [0, 0.75]
ylim = [0, 0.75]
```

```python
# plot using only xlim
fig, ax = plt.subplots(1, 2, figsize=(8, 3))

for a in ax:
    mesh.plotImage(u, ax=a, view="vec", vType="E")
    a.set_xlim(xlim)
    a.set_ylim(ylim)

# resize ax[1]
ax[1].set_xlim([0, 1])
ax[1].set_ylim([0, 1])
```

```python
# plot using only range
fig, ax = plt.subplots(1, 2, figsize=(8, 3))

for a in ax:
    mesh.plotImage(u, ax=a, view="vec", vType="E", range_x=xlim, range_y=ylim)

# resize ax[1]
ax[1].set_xlim([0, 1])
ax[1].set_ylim([0, 1])
```
